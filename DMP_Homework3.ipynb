```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

customer_feedback=pd.read_csv("/content/Customer_Feedback_Data.csv")
product_offering=pd.read_csv("/content/Product_Offering_Data.csv")
transaction=pd.read_csv("/content/Transaction_Data.csv")

# Display Data
print("Customer Feedback Data")
display(customer_feedback.head())

print("\nProduct Offering Data")
display(product_offering.head())

print("\nTransaction Data")
display(transaction.head())
```

```python
# Display missing value count

print("customers:", customer_feedback.shape)
print("Missing value count in Customer Feedback Data")
display(customer_feedback.isnull().sum())

print("products: .", product_offering.shape)
print("\nMissing value count in Product Offering Data")
display(product_offering.isnull().sum())

print("transactions:", transaction.shape)
print("\nMissing value count in Transaction Data")
display(transaction.isnull().sum())
```

```python
# Drop Target Age Group Column
product_offering.drop("Target_Age_Group", axis=1, inplace=True)

display(product_offering.head())
```

```python
# Determining the mean / median for the Satisfaction Scores
mean_score = customer_feedback["Satisfaction_Score"].mean()
median_score = customer_feedback["Satisfaction_Score"].median()

print("Mean:", mean_score)
print("Median:", median_score)

# Imputing missing satisfaction scores
customer_feedback["Satisfaction_Score"] = customer_feedback["Satisfaction_Score"].fillna(median_score)
display(customer_feedback.head())

# Verify imputation
print("\nMissing value count in Customer Feedback Data")
display(customer_feedback.isnull().sum())
```
